# Spam Mail - Machine Learning Prediction

The goal is to use the data provided to build a system that determines if a mail is spam or not

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Data collection and pre-processing**

```python
df = pd.read_csv('/kaggle/input/mail-data/mail_data.csv')

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Replace null values with a null string

df = df.where((pd.notnull(df)),'')

df.head()
```

**Label Encoding**

Spam mail = 0

Ham mail = 1

```python
df.loc[df['Category'] == 'spam', 'Category',] = 0
df.loc[df['Category'] == 'ham', 'Category',] = 1

df.head()
```

**Separate feature(s) and label(s)**

```python
X = df['Message']

Y = df['Category']
```

```python
print(X)
```

```python
print(Y)
```

**Splitting the data into Train and Test data**

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size = 0.75, random_state = 3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Feature extraction**

Transform text data to feature vectors that can be used as input to the Logistic Regression

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# Convert y_train & y_test values as integers
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(X_train_features)
```

**Model training: Logistic Regression**

```python
# Instantiate LogisticRegression
lr = LogisticRegression()
```

```python
# Fit the model
lr.fit(X_train_features, y_train)
```

**Model evaluation: accuracy score**

```python
# prediction on train data
train_data_pred = lr.predict(X_train_features)
train_data_accuracy = accuracy_score(y_train, train_data_pred)

print('Accuracy on training data :', train_data_accuracy)
```

```python
# prediction on test data
test_data_pred = lr.predict(X_test_features)
test_data_accuracy = accuracy_score(y_test, test_data_pred)

print('Accuracy on testing data :', test_data_accuracy)
```

**Building a predictive system**

```python
input_data = ["Had your mobile 11 months or more? U R entitled to Update to the latest colour mobiles with camera for Free! Call The Mobile Update Co FREE on 08002986030"]

# Convert text to feature vectors
input_data_features = feature_extraction.transform(input_data)

# Making prediction
prediction = lr.predict(input_data_features)
print(prediction)

if (prediction[0]==1):
    print('Ham mail')
else:
    print('Spam mail')
```
